Ora suddivido i dati in 3 rami: razza bianca , razza non bianca e razza indefinita e rieffettuo la verifica della camera.

```python
srRace = df4['race'].copy()
```

```python
global i
i = 0
    
def WrapRace(character):  
    global i
    i += 1
    
    if pd.isna(character):
        item = "UnknownRace"
    else:
        if character !='W':
            item = "NotWhite"
        elif character =='W':
            item = "White"
                     
    return item#, print(i)

df4['WhiteAndNot'] = list(map(WrapRace,srRace))
```

```python
dimensions = (8, 8)

fig, ax = plt.subplots(figsize=dimensions)
sns.histplot(ax = ax, data=df4, x='WhiteAndNot', hue="body_camera" )
```

```python
KillNotWithCamera = df4.groupby('WhiteAndNot')["body_camera"].apply(lambda x: (x == False).sum())
KillNotWithCamera
```

6.Cleaning median household income data
===

Utilizzo la libreria "chardet" per riuscire a gestire eventuali caratteri Unicode all'interno del CSV dato che normalmente python utilizza la codifica UTF8.

```python
import chardet

with open(r'MedianHouseholdIncome2015.csv', 'rb') as f:
    result = chardet.detect(f.read()) 
    print(result)
 
dfMedianHouseholdIncome2015 = pd.read_csv('MedianHouseholdIncome2015.csv',encoding=result['encoding'])
dfMedianHouseholdIncome2015["Geographic Area"]=dfMedianHouseholdIncome2015["Geographic Area"].str.strip()
dfMedianHouseholdIncome2015["City"]=dfMedianHouseholdIncome2015["City"].str.strip()
dfMedianHouseholdIncome2015.info()
```

Applicare il metodo strip() è molto importante perchè elimina le tabulazioni e gli spazi bianchi contenuti nei csv evitando che nel merge stringhe con lo stesso contenuto siano riconosciute come diverse.

```python
result
```

```python
df2_MedianHouseholdIncome2015 = dfMedianHouseholdIncome2015.copy()
```

Come convenzione imposto tutte le città con caratteri in minuscolo in modo che nel merge non abbia problemi di matching del case tra i diversi dataframe.

Elimino anche il cdp che dovrebbe essere sia una tag che sta per "census designated place" per lo stesso motivo riportato sopra.

```python
df2_MedianHouseholdIncome2015["City"] = df2_MedianHouseholdIncome2015["City"].str.lower()
```

```python
 
df2_MedianHouseholdIncome2015["City"] = df2_MedianHouseholdIncome2015["City"].str.replace("cdp","")
```

```python
 
df2_MedianHouseholdIncome2015.head(10)
```

```python
dfMedianHouseholdIncome2015.info()
```

Conclusioni:
<ul>
    <li>Time Trend: nel tempo le uccisioni da parte di presunti malviventi da parte della polizia non hanno alcun trend particolare ma oscillano in un range definito.</li>

    <li>Body Camera: solo nel 20% dei casi in media la camera è attiva durante gli interventi della polizia ed è invariante rispetto alla razza dell'aggredito. La credibilità dei poliziotti operando in questo modo è a rischio.</li>
    <li>Povertà e vittime: negli stati dove c'è un forte divario di benessere il numero di vittime è più alto ma non è una relazione di causalità</li>
    <li>Densità di popolazione: il numero di vittime aumenta dove la densità di popolazione è maggiore: il legame è lineare</li>
    <li>Istruzione e densità di razza:
    <ul>
        <li>Bianchi : al crescere della percentuale di razza cresce il numero di diplomati</li>
        <li>Neri : al crescere della percentuale di razza diminuisce in modo moderato il numero di diplomati</li>
        <li>Ispanici : come per i neri</li>
        <li>Nativi : al crescere della percentuale di razza diminuisce in modo decisivo il numero di diplomati</li>
        <li>Asiatici: nessuna correlazione definita</li>
    </ul>

        <li>Neri : al crescere della percentuale di razza la povertà aumenta in modo moderato</li>
        <li>Ispanici : come per i neri</li>
        <li>Nativi : al crescere della percentuale di razza la povertà aumenta in modo decisivo</li>

    </ul>
